```python
import re

def get_processor(x):
    if x=="Intel Core i5" or x=='Intel Core i7' or x=='Intel Core i3':
        return x
    elif x.split()[0] == 'Intel':
        return "Other Intel Processor"
    else:
        return "AMD Processor"

def get_clock_speed(x):
    y = re.findall(r'\d\.?\d+GHz', x)
    if len(y) > 0:
        return y[0].replace('GHz', '')
    else:
        return 0.0
```

```python
df['cpu'] = df['CPU Name'].apply(get_processor)
```

```python
df['clock_speed'] = df['Cpu'].apply(get_clock_speed).astype("float32")
```

```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)

df["first"]= new[0]
df["first"]=df["first"].str.strip()

df["second"]= new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.extract('(\d+)')

df["second"].fillna("0", inplace = True)

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '', regex=True)

df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

df["hdd"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["ssd"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["flash_storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage', 'Memory'],inplace=True)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X = df.drop('price', axis=1)
y = np.log(df['price'])

X['company'] = le.fit_transform(X['company'])
X['type'] = le.fit_transform(X['type'])
X['os'] = le.fit_transform(X['os'])
X['cpu'] = le.fit_transform(X['cpu'])
X['gpu'] = le.fit_transform(X['gpu'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
X_train.sample(10)
```

## Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, ExtraTreesRegressor, GradientBoostingRegressor

svr = SVR(kernel="sigmoid", gamma=1.0)
knr = KNeighborsRegressor(n_neighbors=3)
lr = LinearRegression()
rfr = RandomForestRegressor(n_estimators=50, random_state=101)
abr = AdaBoostRegressor(n_estimators=50, random_state=101)
etr = ExtraTreesRegressor(n_estimators=50, random_state=101)
gbr = GradientBoostingRegressor(n_estimators=50, random_state=101)
```

```python
models = {
    'svr' : svr,
    'knr' : knr,
    'lr' : lr,
    'rfr' : rfr,
    'abr' : abr,
    'etr' : etr,
    'gbr' : gbr,
}
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def train_model(model, model_name):
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)

    print("----------> Model Trained: ", model_name)
    return rmse, mae, r2
```

```python
rmse_scores = []
mae_scores = []
r2_scores = []

for name, model in models.items():
    rmse, mae, r2 = train_model(model, name)
    rmse_scores.append(rmse)
    mae_scores.append(mae)
    r2_scores.append(r2)
```

```python
performance_df = pd.DataFrame({'Algorithm':models.keys(), 'MAE':mae_scores, 'RMSE':rmse_scores, 'R2':r2_scores})
performance_df.sort_values('R2', ascending=False, )
```

## Model Tuning

```python
etr = ExtraTreesRegressor
gbr = GradientBoostingRegressor
```

```python
from sklearn.model_selection import GridSearchCV

models = {
    'etr' : etr,
    'gbr' : gbr,
}

param_grid = {
    'etr': {
        'n_estimators': [50, 100, 200],
        'max_depth': [10, 20, 30, 50],
        'min_samples_split': [2, 5, 10],
    },
    'gbr': {
        'n_estimators': [50, 100, 200],
        'max_depth': [10, 20, 30, 50],
        'min_samples_split': [2, 5, 10],
    },
}

def get_params_string(params):
    params_arr = []
    for key, value in params.items():
        param_string = key + " : " + str(value)
        params_arr.append(param_string)
    return ", ".join(params_arr)

def tune_model(model, model_name):
    grid=GridSearchCV(model(), param_grid[model_name], cv=5, verbose=3)
    grid.fit(X_train, y_train)

    model = grid.best_estimator_
    models[model_name] = model
    
    predictions = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)

    print("----------> Model Trained: ", model_name)
    
    return rmse, mae, r2, get_params_string(grid.best_params_)
```

```python
rmse_scores = []
mae_scores = []
r2_scores = []
params = []

for name, model in models.items():
    rmse, mae, r2, model_params = tune_model(model, name)
    rmse_scores.append(rmse)
    mae_scores.append(mae)
    r2_scores.append(r2)
    params.append(model_params)

performance_df = pd.DataFrame({'Algorithm':models.keys(), 'MAE':mae_scores, 'RMSE':rmse_scores, 'R2':r2_scores, 'Params': params})
```

```python
performance_df.sort_values('R2', ascending=False)
```
