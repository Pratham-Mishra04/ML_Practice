# Popularity Based Recommendation System

```python
book_ratings=ratings.merge(books, on='ISBN')
```

```python
num_rating_series=book_ratings.groupby('Book-Title').count()['Book-Rating']
```

```python
num_rating_df = num_rating_series.reset_index() #Will convert the series to DF
```

```python
num_rating_df.rename(columns={"Book-Rating":"num_ratings"}, inplace=True)
```

```python
num_rating_df
```

```python
book_ratings.info()
```

```python
#groupby and other aggregating functions return a DF only, with
#Book-Title as its index. reset_index() is used to convert Book-Title to a redular column

avg_rating_df = book_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={"Book-Rating": "avg_ratings"}, inplace=True)
```

```python
avg_rating_df
```

```python
# For our popularity based recommender system, we will take top 50 books with highest
# avg ratings, each having minimum 250 ratings
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df = popular_df[popular_df['num_ratings']>250].sort_values('avg_ratings', ascending=False)
```

```python
popular_df
```

```python
# there are multiple ISBN rows for the same Book-Title
popular_df=popular_df.merge(books, on="Book-Title").drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

```python
top_50_df = popular_df.head(50)
```

# Collaborative Filtering Based Recommendation System

```python
#we only only consider movies which have atleast 50 ratings 
#and users who have rated atleast 200 movies
```

```python
# returns objs with the index user-id and values True/False as > 200 is applied
book_ratings.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
users_200ratings_series=book_ratings.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
# filters only the True values from the objs
users_200ratings_series[users_200ratings_series]
```

```python
# returns add the indexes ie the user-ids
users_200ratings_series[users_200ratings_series].index
```

```python
users_200ratings_ids = users_200ratings_series[users_200ratings_series].index
```

```python
filtered_ratings = book_ratings[book_ratings['User-ID'].isin(users_200ratings_ids)]
```

```python
filtered_ratings
```

```python
books_50ratings_series=filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
books_50ratings_titles = books_50ratings_series[books_50ratings_series].index
```

```python
filtered_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(books_50ratings_titles)]
```

```python
filtered_ratings
```

```python
filtered_ratings.info()
```

```python
filtered_ratings.duplicated().sum()
```

```python
# To convert it into a Matrix with Row indexes being Movie Titles, Column indexes being User IDs
# and their corresponding intersection will be the ratings which that user has given to the movie

pt = filtered_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt.shape
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities = cosine_similarity(pt)
```

```python
similarities.shape
```

```python
def recommend(book_title):
    try:
        book_index = np.where(pt.index==book_title)[0][0]
        distances = similarities[book_index]
        book_objs = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
        return [pt.index[i[0]] for i in book_objs]

    except:
        return []
```

```python
recommend("The Notebook")
```

## Saving the Models
